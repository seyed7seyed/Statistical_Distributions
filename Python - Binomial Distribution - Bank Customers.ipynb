# Binomial Distribution

### Importing Libraries

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## Problem Statement
A bank issues credit card to its customers under the scheme of Master card.<Br> Based on past data, it is known that 60% of the customers pay the bill on time.<Br><Br> If a sample of 20 accounts is selected at random from current database:
 - construct the binomial  probability distribution of accounts paying on time.

```python
p, q = 0.6, 0.4

n = 20
k = np.arange(0,n+1)
k
```

```python
binomial = stats.binom.pmf( k, n, p )
binomial
```

```python
# Visualising the plot with pyplot

plt.figure( figsize=(5,3) )
plt.plot(k,binomial,'o-',color='dodgerblue',mfc='deeppink',mec='dodgerblue')
plt.title('Binomial: n=%i , p=%.2f' % (n,p), fontsize=11 )
plt.xlabel('Number of Successes')
plt.ylabel('Probability of Successes')
plt.show()
```
